```python
import matplotlib.pyplot as plt

product=['A','B','C','D','E']
sales = [1000,1500,800,1200,1800]
plt.bar(product,sales,label = "sales 2025",color = "#033952")
# plt.barh(product,sales,label = "sales 2025")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Product Sale comparision")
plt.legend()
plt.show()
```
