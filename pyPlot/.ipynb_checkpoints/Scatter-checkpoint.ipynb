```python
import matplotlib.pyplot as plt

hours_studies = [1,2,3,4,5]
result = [60,65,70,80,90]

plt.scatter(hours_studies,result, color='green',marker='o' , label = 'Class A Student Data' )
plt.scatter(hours_studies,[45,50,65,70,75], color='blue',marker='o' , label = 'Class B Student Data' )
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Relation Between Study time and Score")
plt.legend()
plt.grid(True)
plt.show()
```
